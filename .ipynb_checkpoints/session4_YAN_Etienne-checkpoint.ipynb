```python
path='C:\\Users\\yan_e\\Desktop\\JupyterLab\\data-CPI-2018.xlsx'
cpi = pd.read_excel(path,header=2,index_col=0)
```

```python
for index , isocode in df.isocode.iteritems():
    try :
        ind = cpi[cpi['ISO3']==isocode]['CPI Score 2018']
        rnk = cpi[cpi['ISO3']==isocode]['Rank ']
        df.at[index, 'TI Corruption Perception Index']=ind
        df.at[index, 'TI Corruption Perception Index - Ranking']=rnk
    except ValueError:
        print('[critical]:no data for', isocode)
        continue
```

```python
path='C:\\Users\\yan_e\\Desktop\\JupyterLab\\data-heritage-2019.xls'
herit = pd.read_excel(path,header=0,index_col=0)
```

```python
for index , name in df['Country (according to CN Gov):'].iteritems():
    reducedName= name.split('(')[0]
    reducedName=reducedName.strip()
    if reducedName == 'Thailiand':
        reducedName = 'Thailand'
    elif reducedName == 'The United Arab Emirates':
        reducedName = 'United Arab Emirates'
    elif reducedName =='Brunei' :
        reducedName = 'Brunei Darussalam'
    elif reducedName == 'Myanmar':
        reducedName = 'Burma'
    elif reducedName == 'East Timor':
        reducedName = 'Timor-Leste'
    elif reducedName == 'Kyrgyzstan':
        reducedName = 'Kyrgyz Republic'
    elif reducedName == 'United States of America':
        reducedName = 'United States'
    try: 
        ind = herit[herit['Country Name']==reducedName]['2019 Score']
        rank = herit[herit['Country Name']==reducedName]['World Rank']

        df.at[index, 'HF Index of Economic Freedom (Scale: 1-100)']=ind
        df.at[index, 'Heritage Foundation Index of Economic Freedom (Global Ranking)']=rank
    except ValueError:
        print('[Critical]: no data for ' , reducedName)
```
